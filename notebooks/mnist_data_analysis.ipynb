# MNIST data analysis

### Loading MNIST dataset:

```python
from torchvision import datasets
```

```python
DOWNLOAD = True  # whether to download dataset
```

```python
train_dataset = datasets.MNIST('../data', train=True, download=DOWNLOAD)
test_dataset = datasets.MNIST('../data', train=False, download=DOWNLOAD)
```

### Let's plot some samples:

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))

for i in range(10):
    ax = plt.subplot(2, 5, i + 1)
    ax.set_title(f'Label: {train_dataset[i][1]}')
    ax.imshow(train_dataset[i][0], cmap='gray')
```

### Finally, let's count some statistics:

```python
import numpy as np
```

```python
train_labels = list(map(lambda item: item[1], train_dataset))
test_labels = list(map(lambda item: item[1], test_dataset))
```

```python
print(f'Length of train dataset: {len(train_labels)}')
print(f'Length of test dataset: {len(test_labels)}')
```

```python
plt.figure(figsize=(10, 5))

plt.hist(train_labels, bins=np.linspace(0, 11, num=12)-0.5, edgecolor='black', label='Train dataset')
plt.hist(test_labels, bins=np.linspace(0, 11, num=12)-0.5, edgecolor='black', label='Test dataset')

plt.title("Labels' distribution", fontsize=20)
plt.ylabel('Count', fontsize=16)
plt.xlabel('Digit', fontsize=16)

plt.xticks(range(0, 10))
plt.legend()
plt.show()
```
